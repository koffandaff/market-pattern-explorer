# Model Comparison and Analysis

## Objective

The purpose of this notebook is to compare the performance of all machine learning models trained on the Apple dataset.

The comparison focuses on:

- Predictive performance.
- Generalization ability.
- Feature importance.
- Model complexity.

By comparing multiple approaches, we can understand whether increasing model complexity leads to better market predictions.

```python
# creating a result table 
import pandas as pd

results = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "Random Forest",
        "XGBoost"
    ],

    "Accuracy": [
        0.41,
        0.43,
        0.4008
    ],

    "Precision": [
        0.47,
        0.52,
        0.44
    ],

    "Recall": [
        0.20,
        0.17,
        0.1583
    ]
})

results
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))

plt.bar(
    results["Model"],
    results["Accuracy"]
)

plt.title("Accuracy Comparison")

plt.show()
```

# Key Findings

## 1. Model complexity did not improve performance

More advanced algorithms such as Random Forest and XGBoost did not significantly outperform Logistic Regression.

This suggests that the limiting factor is not the choice of model but the information contained in the features.

---

## 2. Technical indicators alone provide limited predictive power

The feature set consisted entirely of historical price and volume indicators.

External information such as:

- News sentiment.
- Earnings reports.
- Macroeconomic events.
- Market indices.

was not included.

---

## 3. The market signal appears weak

All models achieved accuracy close to random guessing.

This indicates that short-term price direction for Apple stock is difficult to predict using classical technical indicators alone.

---

## 4. Future work

Potential improvements include:

- Hyperparameter tuning.
- Adding sentiment analysis.
- Using longer prediction horizons.
- Incorporating sector and macroeconomic data.
- Exploring LSTM and Transformer architectures.
