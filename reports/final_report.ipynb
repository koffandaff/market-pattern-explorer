# Market Pattern Explorer

## Predicting Short-Term Market Direction Using Classical Machine Learning and Deep Learning

---

### Author

Dhruvil Adroja

---

## Project Goal

The objective of this project was to investigate whether historical market data and technical indicators can be used to predict the short-term direction of Apple stock.

Rather than focusing solely on model performance, this project aimed to understand:

- How market features are engineered.
- How different machine learning models behave.
- Whether model complexity improves predictive power.
- How trading performance differs from classification performance.

---

## Research Question

> Can historical price action and technical indicators predict whether Apple's next trading day will close higher or lower?
-> no it's much deeper than that

---

# Dataset

## Data Source

Historical market data was collected using Yahoo Finance.

Asset:

- Apple Inc. (AAPL)

Collected fields:

- Open
- High
- Low
- Close
- Volume

---

## Dataset Size

- Total observations: 1209 trading days.
- Train/Test split: 80/20.

---

## Prediction Task

The problem was formulated as a binary classification task:

- Class 1 → Tomorrow closes higher.
- Class 0 → Tomorrow closes lower or equal.

Target definition:

Target = 1 if Close(t+1) > Close(t)

Target = 0 otherwise.

Feature engineering became the central part of the project.

---

# Exploratory Observations

Several important observations emerged from visualizing the engineered features.

---

## Noise vs Lag Trade-Off

Short-term indicators:

- React quickly.
- Capture recent market movements.
- Contain significant noise.

Examples:

- Return
- Change
- Momentum

Long-term indicators:

- Smooth market fluctuations.
- Reveal larger trends.
- React slowly.

Examples:

- MA_50
- Volatility

---

## Insights

- MA_10 closely followed the price.
- MA_50 reacted much more slowly.
- Returns were highly noisy.
- Trend-following indicators appeared more stable.

These observations motivated the selection of features used by the models.


---

# Models Investigated

The project explored four different approaches.

---

## Logistic Regression

Purpose:

- Build a simple and interpretable baseline.

Assumption:

- Market relationships are approximately linear.

---

## Random Forest

Purpose:

- Capture non-linear interactions.

Advantages:

- Robust to noise.
- Handles feature interactions automatically.

---

## XGBoost

Purpose:

- Improve performance using boosting.

Advantages:

- Sequential error correction.
- Regularization.

---

## LSTM

Purpose:

- Learn temporal dependencies.

Advantages:

- Uses sequences instead of independent rows.
- Can model long-term patterns.


# Model Performance

| Model | Accuracy | Precision | Recall |
|--------|--------|--------|--------|
| Logistic Regression | 41.0% | 47.0% | 20.0% |
| Random Forest | 43.0% | 52.0% | 17.0% |
| XGBoost | 40.1% | 44.0% | 15.8% |
| LSTM | 45.8% | 53.5% | 40.0% |

---

## Observations

- Random Forest slightly outperformed Logistic Regression.
- XGBoost did not improve performance.
- LSTM achieved the best overall results.
- Increasing model complexity alone did not solve the problem.

---

# Backtesting Results

Machine learning performance did not directly translate into trading performance.

The Buy-and-Hold strategy significantly outperformed all machine learning strategies.

---

## Sharpe Ratios

| Model | Sharpe Ratio |
|--------|--------|
| Logistic Regression | -1.94 |
| Random Forest | 0.21 |
| XGBoost | 0.13 |

---

## Key Observation

Even though Random Forest achieved better classification metrics, it failed to outperform simply holding Apple stock.

This highlighted an important lesson:

> Accuracy alone does not imply profitability.

---

# Failures and Lessons Learned

Several assumptions made at the beginning of the project proved incorrect.

---

## Assumption 1

"More complex models will automatically perform better."

Result:

False.

---

## Assumption 2

"Technical indicators contain enough information."

Result:

Partially false.

---

## Assumption 3

"Accuracy is the most important metric."

Result:

False.

Backtesting revealed that portfolio performance depends on:

- Risk.
- Market participation.
- Drawdowns.
- Missed opportunities.

---

## Major Lesson

Feature engineering and evaluation methodology matter more than model complexity.

## Final Conclusion

This project demonstrated that predicting short-term stock movements using historical prices alone is extremely difficult.

The primary contribution of the project was not achieving high accuracy, but developing a deeper understanding of:

- Financial feature engineering.
- Machine learning workflows.
- Time-series modeling.
- Backtesting.
- The difference between predictive performance and real-world profitability.

The project evolved from a modeling exercise into an investigation of market efficiency and the limitations of data-driven prediction.
